```python
# Q1 Which keyword is used to create a function? Create a function to return a list of odd numbers in the 
# range of 1 to 25

# def keyword is used to create a function in python

def onum():
  a=[]
  for i in range(1,26):
    if i%2 !=0:
      a.append(i)
  return a 

onum()
```

```python
# Q2--Why *args and **kwargs is used in some functions? Create a function each for *args and **kwargs to 
# demonstrate their use.


# args help in taking multiple arguments , while kwargs help in defining key value pair in a function

# *args

def marg(*args):
  print(args[0]+args[1]*args[2])

marg(1,2,3)
```

```python

print("Kwargs example")

def keyarg(**kwargs):
  for i in  kwargs:
    print(i,kwargs[i])

keyarg(
    name="harsh",
    age=21,
    sex="Male"
)
```

```python
# Generator is basically used to create iterators
# yield keyword makes entire function iterator 

def square(n):
  for i in range(n):
    yield  i*i

a=square(3)

next(a)
next(a)
```

```python
# yield will return with respect to every iteration not just one and saves local variable

for i in square(5):
  print(i)
```
