```python
dict1 ={'Name':["Harsh",'Aadhya','Krisha','Arnab',"rashi"],
        'Marks':[99,89,99,1,11],
        'Gender':['Male','Female','Female','Male','Female'],
        'Date':['2023-01-03','2023-02-22','2023-01-23','2023-03-12','2023-01-13']
        }
df4=pd.DataFrame(dict1)
df4
```

```python
def datecheck(x):
  if x[5:7]=="01":
    return x

df4["Jan_Data"]=df4['Date'].apply(datecheck)
df4
```
