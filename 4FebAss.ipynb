```python
# Write a python program using reduce function to compute the product of a list containing numbers 
# from 1 to 25.

from functools import reduce

l2=[]
for i in range(1,26):
  l2.append(i)
print(l2)
```

```python
reduce(lambda x,y:x*y,l2)
```

```python
# Write a python program to filter the numbers in a given list that are divisible by 2 and 3 using the 
# filter function.

list_1 = [2, 3, 6, 9, 27, 60, 90, 120, 55, 46]

list(filter(lambda x:x%6==0,list_1))
```

```python
l1=['python', 'php', 'aba', 'radar', 'level']

for i in l1:
  if i[:]==i[::-1]:
    print(i,"is a palindrome")
  else:
    print(i,"is not a palindrome")
```

```python
list(map(lambda x:x[:]==x[::-1],l1))
```

```python
# Checking if its a palindrom
# checked by character wise

str1= input("Enter a string ")

a1=[]
for i in str1:
  a1.append(i)
print(a1)

print("-----------------------")
a2=[]
for i in str1[::-1]:
  a2.append(i)
print(a2)

print("---------------------")

def check(l1,l2):
  if l1==l2:
    print("its a palindrom")
  else:
    print("Not  a palindrom")

check(a1,a2)
```
