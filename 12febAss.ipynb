```python
try:
  with open("Harsh.txt","r") as f:
    print(f.read())

except Exception as FileNotFoundError:
  print("No such file found -",FileNotFoundError)
finally:
  print("I am always printed")

print("Even after error print statement works")
```

```python
class customclass(Exception):
  def __init__(self,msg):
    self.msg=msg
```

```python
def check(age):
  if age<0:
    try:
      raise customclass("Age cant be -ve")
    except Exception as e:
      print("Sorry",e)
  elif age>120:
    try:
      raise customclass("Age too high")
    except Exception as e:
      print("Sorry",e)
  else:
    print("Age is apt")
```

```python
age1=int(input("Enter ur age "))
a1=check(age1)

```
