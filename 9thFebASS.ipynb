```python
class Myclass:
  def __init__(self,value):
      self._value=value

  def show(self):
     print(f"Value is {self._value}")

# gettters
  @property
  def a(self):
    return 10*self._value

# setters
  @a.setter
  def a(self,new_value):
     self._value=new_value/10


```

```python
# Method overriding means when we want to redefine parent class wrt child class

class shape:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def area(self):
    return self.x*self.y

sq=shape(10,12)
sq.area()
```

```python
class Circle(shape):
  def __init__(self,radius):
    self.radius=radius
    super().__init__(radius,radius)

  def area(self):
    return 3.14*super().area()

c=Circle(5)
c.area()
```

```python
# What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

class A():
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def m1(self):
    print(f"Method  of class A is called and name is {self.name} and age is {self.age}")

class B():
  def __init__(self,sex):
    self.sex=sex
  def m2(self):
    print(f"Method  is of class B is called and  sex is {self.sex}")

class C(A,B):
  
  def __init__(self,name,age):
    super().__init__(name,age)
  
  def m3(self):
    print(f"Method of class C is called and name is {self.name} and age is {self.age} and sex is ")

c2=C("Deepti",20)
c2.m3()
```

```python
''' Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class. 
Create a method named seating_capacity which takes capacity as an argument and returns the name of 
the vehicle and its seating capacity.'''


class vehicle:
  def __init__(self,name,speed,avg):
    self.name=name
    self.speed=speed
    self.avg=avg
  
class car(vehicle):
  def __init__(self,name,speed,avg,capacity):
    super().__init__(name,speed,avg)
    self.capacity=capacity
  
  def det(self):
    print(f"The name of car is {self.name} and seating capacity is {self.capacity}")

x1=car("Ford",100,50,4)
x1.det()
```
