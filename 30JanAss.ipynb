```python
# Accept percentage of user and display the grades

a=int(input("Enter ur percentage"))

if a>90:
  print("Grade A")
elif a>80 and a<=90:
  print("Grade B")
elif a>=60 and a<=80:
  print("Grade c")
else:
  print("Grade D")
```

Q6

```python
num_spaces_wanted = 0
while num_spaces_wanted < 7:
    print('#', end='')
    num_spaces_printed = 0
    while num_spaces_printed < num_spaces_wanted:
        print(' ', end='')
        num_spaces_printed += 1
    print('#')
    num_spaces_wanted += 1
```
