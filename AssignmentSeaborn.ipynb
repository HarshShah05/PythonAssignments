```python
#  Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = 
# "timepoint" and y = "signal" for different events and regions

import seaborn as sns

fmri=sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot(data=fmri,x="timepoint",y="signal")
```

```python
#  Use the "flights" dataset from seaborn to plot a heatmap.
# always we need to pivot dataset in order to plot heatmaps---SNS documentation

flights=sns.load_dataset("flights")
flights
```

```python
flights1=flights.pivot_table(index="month",columns="year", values="passengers")
flights1
```

```python
sns.set(rc={"figure.figsize":(20, 5)})
sns.heatmap(flights1,cmap="Blues",linecolor="white",linewidth=2)
```
