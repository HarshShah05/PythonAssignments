```python
# What is inheritance? Give an example for each type of inheritance
# inheritance is basically used when parent class inherits properties of child class
# there are various inheritance like multiple,multilevel,normal

# Normal inheritance
class employee():
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def show(self):
    print(f"Name is {self.name} and age is {self.age}")

class prog(employee):
  def lang(self):
    print("The language he speaks is gujrati")

# x1=employee("David",10)
# x1.show()

x2=prog("Sam",31)
x2.show()
x2.lang()
```

Multilevel inheritance

```python
class A():
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def m1(self):
    print(f"Method  of class A is called and name is {self.name} and age is {self.age}")

class B(A):
  def __init__(self,name,age,sex):
    super().__init__(name,age)
    self.sex=sex
  def m2(self):
    print(f"Method  is of class B is called and name is {self.name} and age is {self.age} and sex is {self.sex}")

class C(B):
  def __init__(self,name,age,sex,id):
    super().__init__(name,age,sex)
    self.id=id
  
  def m3(self):
    print(f"Method of class C is called and name is {self.name} and age is {self.age} and sex is {self.sex} and id is {self.id}")

print("Object of A only can acces method of A")
a1=A("Harsh",21)
a1.m1()

print("--------------------------------------------------------------")

print("Object of B  can acces method of A and B\n")
b1=B("Heta",23,"Female")

print("Accesing method of A")
b1.m1()
print("\n")
print("Accesing method of B")
b1.m2()


print("---------------------------------------------------------------------------------")

print("Object of C  can acces method of A and B and C")
c1=C("Rashi",25,"Female",100)

print("Accesing method of A")
c1.m1()
print("\n")
print("Accesing method of B")
c1.m2()
print("\n")
print("Accesing method of C")
c1.m3()
```

Multiple inheritance


```python
class A():
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def m1(self):
    print(f"Method  of class A is called and name is {self.name} and age is {self.age}")

class B():
  def m2(self):
    print("i am b method")
   

class C(A,B):
  def m3(self):
    print(f"Method of class C is called and name is {self.name} and age is {self.age} ")

c2=C("Deepti",20)

print("Accessing C method")
c2.m3()

print("\n")
print("Accessing B method")
c2.m2()
print("\n")
print("Accessing A method")
c2.m1()




```
