```python
# Write a Python program that reads a CSV file containing sales data for different products and 
# visualizes the data using a stacked bar chart to show the sales of each product category over time. The 
# program should prompt the user to enter the file path and display the chart.

path=input("Enter path:")
df12=pd.read_csv(path)
df12
```

```python
import matplotlib.pyplot as plt

plt.bar(df12.Year,df12.Bat,label="Bat")
plt.bar(df12.Year,df12.Ball,label="Ball")
plt.bar(df12.Year,df12.Stumps,label="Stumps")
plt.bar(df12.Year,df12.Gloves,label="Gloves")
plt.xlabel("Years")
plt.ylabel("Products sales")
plt.title("Product sales over time")
plt.xticks(df12.Year[::1])
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

for i in df12:
  if i!="Year":
    plt.bar(df12.Year,df12[i])
plt.xlabel("Years")
plt.ylabel("Products sales")
plt.title("Product sales over time")
plt.xticks(df12.Year[::1])

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(df12.Year,df12.Bat,label="Bat",marker=".")
plt.plot(df12.Year,df12.Ball,label="Ball",marker="*")
plt.plot(df12.Year,df12.Stumps,label="Stumps",marker="o")
plt.plot(df12.Year,df12.Gloves,label="Gloves",marker="D")
plt.xlabel("Years")
plt.ylabel("Products sales")
plt.title("Product sales over time")
plt.xticks(df12.Year[::1])
plt.legend()

plt.show()
```
